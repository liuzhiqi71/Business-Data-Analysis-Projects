```python
import pandas as pd
import matplotlib.pyplot as plt

# 创建数据表
data = [
    ["Home Depot", 42.27, 25.85, 1662630],
    ["Microsoft", 37.8, 26.00, 622310],
    ["UnitedHealth", 39.2, 23.92, 447861],
    ["SunAmerica", 36.5, 25.73, 426109],
    ["Adobe", 37.38, 23.93, 304043],
    ["Amazon", 26.63, 31.78, 155451],
    ["Nvidia", 24.94, 33.38, 131754],
    ["PLDT", 32.05, 25.07, 129945],
    ["Plenum Publishing", 25.6, 32.09, 124243],
    ["Aaon", 32.99, 24.03, 121744],
    ["NVR", 30.11, 24.46, 72669],
    ["Pool Corp", 28.21, 25.27, 57579],
    ["Time Warner", 26.24, 27.11, 54160],
    ["Express Scripts", 26.53, 25.97, 45718],
    ["Axon Enterprise", 22.56, 31.13, 45223],
    ["Netflix", 21.6, 32.06, 40613],
    ["Cognizant Technology", 25.53, 26.40, 39592],
    ["Manor Care", 25.78, 26.08, 39324],
    ["National Computer Systems", 27.77, 23.63, 36164],
    ["Biomet", 24.77, 26.73, 35345],
    ["Lin Broadcasting", 22.8, 28.41, 29926],
    ["Sperry Corp", 22.16, 28.82, 27367],
    ["Computer Data Systems", 22.38, 28.15, 25748],
    ["Pioneer Group", 21.53, 27.98, 20266],
    ["Logicon", 24.63, 23.96, 19839],
    ["Keurig Green Mountain", 22.45, 26.30, 18902],
    ["Intuitive Surgical", 23.55, 24.31, 16812],
    ["Metromedia", 22, 24.88, 13269],
    ["Eckerd Jack Corp", 22.27, 24.40, 12932],
    ["Neutrogena", 21.46, 25.17, 12370],
]

df = pd.DataFrame(data, columns=["Company", "Years", "Annualised Compound Return", "Cumulative Return"])

# 归一化气泡大小，以防止过大
bubble_size = df["Cumulative Return"] / df["Cumulative Return"].max() * 1000

# 绘制气泡图
plt.figure(figsize=(15, 10))
scatter = plt.scatter(
    df["Years"], df["Annualised Compound Return"], 
    s=bubble_size, alpha=0.6, edgecolors="#de2d26", color="#de2d26"
)

# 添加标签
for i, txt in enumerate(df["Company"]):
    plt.annotate(txt, (df["Years"][i], df["Annualised Compound Return"][i]), 
                 fontsize=9, alpha=0.85, color="black")

# 设置标题和标签
plt.xlabel("Years", fontsize=12, color="black")
plt.ylabel("Annualised Compound Return (%)", fontsize=12, color="black")
plt.title("Bubble Chart of Company Stock Performance, at least 20 years history", fontsize=14, color="black")

# 调整刻度字体
plt.xticks(fontsize=10, color="black")
plt.yticks(fontsize=10, color="black")

# 定义保存路径
output_path = "best_return.png"

# 保存图像
plt.savefig(output_path, dpi=300, bbox_inches="tight")

# 显示图表
plt.show()
```
