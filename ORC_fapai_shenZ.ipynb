```python
import os
import requests
from bs4 import BeautifulSoup

# 公众号文章主页 URL（替换成你的公众号主页）
main_url = "https://mp.weixin.qq.com/s/iZehOpcBelD9eQnIRLnTJg"

# 设置请求头，伪装成浏览器
session = requests.Session()
session.headers.update({
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/110.0.5481.77 Safari/537.36"
    ),
    "Referer": "https://mp.weixin.qq.com/"
})

# 你要搜索的关键词（仅提取包含该关键词的链接）
keyword = "1月深圳法拍房汇总"  # 例如 "Python"、"AI"、"新闻"

# 下载图片存储目录
save_dir = "wechat_images"
os.makedirs(save_dir, exist_ok=True)

def get_soup(url):
    """ 获取网页内容并解析成 BeautifulSoup """
    try:
        response = session.get(url, timeout=10)
        response.raise_for_status()
        return BeautifulSoup(response.text, "html.parser")
    except requests.RequestException as e:
        print(f"请求失败: {url}, 错误: {e}")
        return None

def find_links_with_keyword(main_url, keyword):
    """ 从公众号文章页面找到包含特定关键词的链接 """
    soup = get_soup(main_url)
    if not soup:
        return []

    links = []
    for a_tag in soup.find_all("a", href=True):
        href = a_tag["href"]
        text = a_tag.get_text(strip=True)

        # 检查链接文本是否包含关键词
        if keyword in text or keyword in href:
            links.append(href)

    return links

def download_images_from_url(url):
    """ 从指定文章链接下载图片 """
    soup = get_soup(url)
    if not soup:
        return

    # 提取所有图片链接
    img_tags = soup.find_all("img")
    img_urls = [img["data-src"] for img in img_tags if "data-src" in img.attrs]

    for idx, img_url in enumerate(img_urls):
        try:
            img_response = session.get(img_url, timeout=10)
            img_response.raise_for_status()
            img_path = os.path.join(save_dir, f"{url.split('/')[-1]}_image_{idx}.jpg")

            with open(img_path, "wb") as f:
                f.write(img_response.content)

            print(f"已下载: {img_path}")
        except requests.RequestException as e:
            print(f"下载失败: {img_url}, 错误: {e}")

# **运行流程**
if __name__ == "__main__":
    # **Step 1: 获取符合条件的链接**
    article_links = find_links_with_keyword(main_url, keyword)
    print(f"找到 {len(article_links)} 个包含关键词 '{keyword}' 的链接")

    # **Step 2: 遍历这些链接，提取图片**
    for article_url in article_links:
        print(f"正在处理文章: {article_url}")
        download_images_from_url(article_url)

    print("所有符合条件的图片下载完成！")
```

```python
import os
import pytesseract
from PIL import Image
import csv
import re

# 图片所在目录（请根据实际情况修改路径）
image_directory = "/Users/weiliu/Python-Practices from Mine/Data Science/wechat_images"

# 输出 CSV 文件名
output_csv = "bad_loan_output.csv"

def extract_relevant(text):
    """
    提取文本中的中文字符、数字以及日期常用符号（-、/、.、:）
    """
    # 正则模式：匹配中文、数字和 - / . :
    pattern = r'[\u4e00-\u9fff0-9\-/\.:]+'
    return ''.join(re.findall(pattern, text))

def ocr_image(image_path):
    """
    对图片进行 OCR，使用中文+英文语言包
    """
    # 使用 chi_sim+eng 以确保数字及日期字符能被识别
    return pytesseract.image_to_string(Image.open(image_path), lang="chi_sim+eng")

# 打开 CSV 文件以写入（注意编码设置保证中文正确）
with open(output_csv, mode="w", newline='', encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    
    # 遍历目录中所有图片
    for filename in os.listdir(image_directory):
        if filename.lower().endswith((".png", ".jpg", ".jpeg", ".tiff", ".bmp", ".gif")):
            image_path = os.path.join(image_directory, filename)
            print(f"Processing {image_path}")
            
            # OCR 提取整张图片文本
            text = ocr_image(image_path)
            # 按行拆分
            lines = text.split('\n')
            
            # 逐行处理
            for line in lines:
                # 提取中文、数字及日期符号
                relevant_text = extract_relevant(line)
                if relevant_text:  # 如果提取出内容
                    # 每一行写入 CSV 一行记录（可根据需要调整分列方式）
                    writer.writerow([relevant_text])

print("CSV file created successfully.")
```

```python
import re
import csv

# 输入的原始 CSV 文件（每行一条OCR提取的字符串）
input_csv = '/Users/weiliu/Python-Practices from Mine/Data Science/bad_loan_output.csv'
# 输出解析后的 CSV 文件
output_csv = "parsed_output.csv"

# 定义目标列名
columns = ["开拍", "区域", "状态", "标的物名称", "用途", "面积", "起拍价", "起拍单价", "保证金"]

# 正则说明：
# 1. ^(\d+/\d+)           —— 开拍日期，如 "2/18"
# 2. ((?:(?!一拍|二拍|变卖).)+)? —— 可选区域（匹配直到遇到状态关键词，注意：区域可能为空）
# 3. (一拍|二拍|变卖)      —— 状态（固定为“一拍”、“二拍”或“变卖”）
# 4. (.*?)                —— 标的物名称（非贪婪匹配，尽可能少地捕获字符）
# 5. ((?:住宅|商住楼))?    —— 可选用途（“住宅”或“商住楼”）
# 6. (\d+(?:\.\d{2})?)     —— 面积（数字，允许带2位小数，也允许整数）
# 7. (\d+(?:\.\d{2})?)     —— 起拍价（同上）
# 8. (\d+(?:\.\d{2})?)$    —— 起拍单价（同上）
pattern = re.compile(
    r'^(\d+/\d+)'                       # 开拍日期
    r'((?:(?!一拍|二拍|变卖).)+)?'        # 可选区域（不包含状态关键词）
    r'(一拍|二拍|变卖)'                  # 状态
    r'(.*?)'                           # 标的物名称（非贪婪）
    r'((?:住宅|商住楼))?'               # 可选用途
    r'(\d+(?:\.\d{2})?)'                # 面积
    r'(\d+(?:\.\d{2})?)'                # 起拍价
    r'(\d+(?:\.\d{2})?)$'               # 起拍单价
)

with open(input_csv, mode="r", encoding="utf-8-sig") as infile, \
     open(output_csv, mode="w", newline='', encoding="utf-8-sig") as outfile:
    
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    
    # 写入表头
    writer.writerow(columns)
    
    for row in reader:
        if not row:
            continue
        text = row[0].strip()
        m = pattern.match(text)
        if m:
            open_date = m.group(1)                          # 开拍日期
            region = m.group(2) if m.group(2) else ""         # 区域（可选）
            status = m.group(3)                             # 状态
            property_name = m.group(4).strip()              # 标的物名称
            usage = m.group(5) if m.group(5) else ""          # 用途（可选）
            area = m.group(6)
            starting_price = m.group(7)
            unit_price = m.group(8)
            deposit = ""  # 保证金字段暂无数据
            
            writer.writerow([open_date, region, status, property_name, usage, area, starting_price, unit_price, deposit])
        else:
            print(f"无法解析：{text}")

print("CSV 文件处理完成！")
```
